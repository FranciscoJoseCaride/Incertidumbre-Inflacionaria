```python
import pandas as pd
```

```python
data = pd.read_excel('IPC.xlsx') #importo los datos
IPC = pd.DataFrame(data) #la convierto en un dataframe
IPC=IPC.set_index('Periodo') #marco la columna de Periodo como la key del dataframe
IPC_1er_diff = IPC.diff(1) #diferencio la serie
IPC_1er_diff = IPC_1er_diff.dropna() #elimino el Nan que queda después de diferenciar la serie
IPC_1er_diff.rename(columns={'Periodo': 'Periodo', 'IPC_Var_Anual_Nivel': 'IPC_Var_Anual_1erDif'}, inplace=True) #cambio el nombre de las columnas
IPC_1er_diff = IPC_1er_diff.reset_index() #agrego una nueva columna como indice
promedio_1er_diff = IPC_1er_diff['IPC_Var_Anual_1erDif'].mean() #calculo el promedio general
IPC_1er_diff['Periodo'] = IPC_1er_diff['Periodo'].apply(lambda x: x.strftime('%B')) #cambio el formato de los peridos para que sean solo meses
promedios_mensuales_IPC_1er_diff = IPC_1er_diff.groupby('Periodo').mean() #saco el promedio mensual
promedios_mensuales_IPC_1er_diff['coef_estac'] = promedios_mensuales_IPC_1er_diff['IPC_Var_Anual_1erDif'] - promedio_1er_diff #calculo el coeficiente de estacionalidad restando el promedio de cada mes con el promedio general
promedios_mensuales_IPC_1er_diff=promedios_mensuales_IPC_1er_diff.drop(['IPC_Var_Anual_1erDif'], axis=1) #elimino la columna de promedio mensual para quedarme con un df con solo los coeficientes de cada mes
```
